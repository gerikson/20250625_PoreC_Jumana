```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.tsv' with the actual path to your file
file_path = '/data/alhajabed/FlyHiC/PoreC/20250625_balancedMcools/20250709_PnMs_merged_balanced.20kb..eigenvectors.txt'

# Read the tab-separated file
df = pd.read_csv(file_path, sep='\t')

# Drop rows with missing E1 values
df = df.dropna(subset=["E1"])

# Compute midpoints of each bin
midpoints = (df["start"] + df["end"]) / 2

# Plot
plt.figure(figsize=(10, 5))
plt.plot(midpoints, df["E1"], marker='o', linestyle='-')
plt.xlabel("Genomic Position (bp)")
plt.ylabel("E1 Eigenvector")
plt.title("Hi-C Eigenvector (E1) Signal")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file_path = '/data/alhajabed/FlyHiC/PoreC/20250625_balancedMcools/20250709_PnMs_merged_balanced.20kb.top100.eigenvectors.txt'
file_path_439 = '/data/alhajabed/FlyHiC/PoreC/20250625_balancedMcools/20250709_PnMs_merged_balanced.20kb.top100.439.eigenvectors.txt'

# Read the data
df = pd.read_csv(file_path, sep='\t').dropna(subset=["E1"])
df_439 = pd.read_csv(file_path_439, sep='\t').dropna(subset=["E1"])

# Compute midpoints
midpoints = (df["start"] + df["end"]) / 2
midpoints_439 = (df_439["start"] + df_439["end"]) / 2

# Plot both eigenvectors
plt.figure(figsize=(12, 6))
plt.plot(midpoints, df["E1"], marker='o', linestyle='-', label='2L_057')
plt.plot(midpoints_439, df_439["E1"], marker='s', linestyle='--', label='2L_439')

# Plot formatting
plt.xlabel("Genomic Position (bp)")
plt.ylabel("E1 Eigenvector")
plt.title("Hi-C Eigenvector (E1) Signal Comparison")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
